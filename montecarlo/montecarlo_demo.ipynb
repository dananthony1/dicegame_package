```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from montecarlo import Die
from montecarlo import Game
from montecarlo import Analyzer
```

```python
# Scenario 1: A 2-headed Coin
FairCoin = Die(faces=np.array(['H','T']))
UnfairCoin = Die(faces=np.array(['H','T']))
```

```python
UnfairCoin.change_weight_single_side('H',5)
```

```python
FairList = [FairCoin]
FairGame = Game(FairList)
```

```python
FairGame.play(1000)
```

```python
UnfairList = [FairCoin, UnfairCoin, UnfairCoin]
UnfairGame = Game(UnfairList)
```

```python
UnfairGame.play(1000)
```

```python
AF = Analyzer(FairGame)
AU = Analyzer(UnfairGame)
```

```python
print(AF.jackpot())
print(AU.jackpot())

relativeFair = AF.jackpot()/1000
relativeUnfair = AU.jackpot()/1000
data = [relativeFair,relativeUnfair]
print(relativeFair)
print(relativeUnfair)
```

```python
fig = plt.figure()
s = fig.add_subplot(111)
s.bar([1, 2], [relativeFair, relativeUnfair], width=1)
s.set_xlim(0.5, 3.5)
```
